```python
import os
import keras
import numpy as np
from PIL import Image
import cv2  
import tensorflow as tf
#import numpy 
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.models import load_model 

def apply_color_mask(hsv,img,low,high):
    # Apply color mask to image
    mask = cv2.inRange(hsv, low, high)
    res = cv2.bitwise_and(img,img, mask= mask)
    return res
os.environ["CUDA_VISIBLE_DEVICES"] = "1"
config = tf.compat.v1.ConfigProto()
config.gpu_options.allow_growth = True
session = tf.compat.v1.Session(config=config)
model=load_model('cnn.h5')
```

```python
#import cv2

# 讀取圖檔
im = cv2.imread('diease/01cff44f-9564-42f7-9a29-3daa487b306a___RS_L.Scorch 1333.JPG')
im=cv2.resize(im,(128,128), interpolation = cv2.INTER_CUBIC)
im_MSF= cv2.pyrMeanShiftFiltering(im, 5, 5, None, 2)
plt.imshow(cv2.cvtColor(im_MSF,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 建立 Selective Search 分割器
ss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation()

# 設定要進行分割的圖形
ss.setBaseImage(im_MSF)

# 使用快速模式（精準度較差）
#ss.switchToSelectiveSearchFast()

# 使用精準模式（速度較慢）
ss.switchToSelectiveSearchQuality()

# 執行 Selective Search 分割
rects = ss.process()

print('候選區域總數量： {}'.format(len(rects)))

# 要顯示的候選區域數量
numShowRects = 1000

# 每次增加或減少顯示的候選區域數量
increment = 50




  # 以迴圈處理每一個候選區域
for i, rect in enumerate(rects):
  # 以方框標示候選區域
    
    if (i < numShowRects):
        imOut = im.copy()
        x, y, w, h = rect
        imrect=im_MSF[y:y+h,x:x+w]
        imrect=cv2.resize(imrect,(128,128), interpolation = cv2.INTER_CUBIC)
        plt.subplot(1,2,1)
        plt.imshow(cv2.cvtColor(imrect,cv2.COLOR_BGR2RGB))
        imrect=imrect.reshape(-1,128,128,3)
        pred=model.predict(imrect)
        if(np.argmax(pred)==0):
            
            cv2.rectangle(imOut, (x, y), (x+w, y+h), (0, 0, 255), 1, cv2.LINE_AA)
            plt.subplot(1,2,2)
            plt.imshow(cv2.cvtColor(imOut,cv2.COLOR_BGR2RGB))
        plt.show()
    else:
        break
```
